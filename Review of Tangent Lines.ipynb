```python
import numpy as np
import matplotlib.pyplot as plt
```

## Section 1: What is a Tangent Line?

A tangent line is used to calculate the slope of a function at a specific point. This is different from a secant point, since only one point is used for a tangent line, while a secant line uses two.

Some notation: a horizontal tangent means that the derivative at the point is 0. One example is at the vertex of a parabola, as shown below.

```python
# Parabola Vertex and Horizontal Tangent (notice that the orange line is completely flat)

t = np.linspace(-1,1,300)
y = t**2
y1 = 0*t
plt.plot(t,y)
plt.plot(t,y1)
plt.scatter(0,0,s=70,facecolor='k',edgecolor='k')
plt.show()
```

Our equation for a tangent line at the point $a$ for a function $f(x)$ is: $y = f'(a)x + b$.

## Section 2: Examples

Let's find the tangent line of $f(x) = \cos(x) - \sin(2x)$ at $x = 0$. We can see that $f(0) = 1$, We can also see that $f'(x) = -\sin(x) - 2\cos(2x)$, and $f'(0) = -\sin(0) - 2\cos(0) = -2$, which gives us our slope at $x=0$.

We can put this into our tangent line definition, and we get that $y = -2x+b$. Now, using the point $(0,1)$ on the function, we can go and find $b$: our y-intercept. We see that $1 = -2(0)+b$, so $b=1$. Thus, our tangent line for $f(x)$ at $x=0$ is $y=-2x+1$. Below is a graph for visualization.

```python
# f(x) and Tangent Line
t = np.linspace(-1.5,1.5,400)
y = np.cos(t) - np.sin(2*t)
y1 = 1-2*t
plt.plot(t,y)
plt.plot(t,y1)
plt.scatter(0,1,s=70,facecolor='k',edgecolor='k')
plt.show()
```

Notice how the tangent line lies parallel to the curve.

Let's do another tangent line, say $f(x) = e^{-2x} + (2x-3)^3$ at $x=0$. We see that $f(0) = -26$, and $f'(x) = -2e^{-2x} + 6(2x-3)^2$. Now, we get that $f'(0) = -2 + 54 = 52$, so our tangent line is $y = 52x + b$. We can use the point $(0,-26)$ to find $b$. We get that $-26 = 0 + b$, so $b = -26$. Thus, the tangent line is $y = 52x-26$. Another graph is attached to show the tangent line and graph.

```python
# f(x) and Tangent Line
t = np.linspace(-0.5,1.5,400)
y = np.exp(-2*t) + (2*t - 3)**3
y1 = 52*t - 26
plt.plot(t,y)
plt.plot(t,y1)
plt.scatter(0,-26,s=70,facecolor='k',edgecolor='k')
plt.ylim(-50,0)
plt.show()
```
