```python
import numpy as np
import matplotlib.pyplot as plt
```

## Section 1: When is a Function Continuous?

A function $f(x)$ is continuous when its functional value at $x=a$ is equal to its limit value at $x=a$. That is, 

$$\lim_{x \to a}f(x) = f(a)$$

If this condition is not satisfied, then the function is not continuous.

## Section 2: Now When is a Function Not Continuous?

There are four types of discontinuities.

1. Jump Discontinuities: these occur when a function suddenly jumps from one value to another. The limits from the left and right differ, thus, the limit does not exist. A graph is below, where the discontinuity is at $x=1$.

```python
# Jump Discontinuity Example
x1 = np.linspace(-3,1,400)
x2 = np.linspace(1.08,3,400)
y1 = 2**x1
y2 = 1-x2**2

plt.plot(x1,y1,color='r')
plt.plot(x2,y2,color='r')
plt.scatter(1, 2,s=70, facecolor='none', edgecolor='k')
plt.scatter(1, 0,s=70, facecolor='k', edgecolor='k')
plt.show()
```

2. Infinite Discontinuities: these occur when a function goes to $+ \infty$ or $- \infty$. This can occur on either the left side, the right side, or both sides, even when $f(x)$ is defined before the point of discontinuity. Some graphs are below that illustrate infinite discontinuities.

```python
# Infinite Discontinuity Example 1
x1 = np.linspace(-2,0,400)
x2 = np.linspace(0.01,2,400)
y1 = x1
y2 = np.log(x2)

plt.plot(x1,y1,color='r')
plt.plot(x2,y2,color='r')
plt.scatter(0,0,s=70,facecolor='k',edgecolor='k')
plt.show()
```

```python
# Infinite Discontinuity Example 2
x1 = np.linspace(-2,-0.01,400)
x2 = np.linspace(0.001,2,400)
y1 = 1/(30*x1)
y2 = -np.log(x2)

plt.plot(x1,y1,color='r')
plt.plot(x2,y2,color='r')
plt.show()
```

3. Hole Discontinuities: these occur when there is a point removed from the function, and in its place is a hole. However, the limit still exists at the discon- tinuous point. Basically, the point is completely gone from the graph.

```python
# Hole Discontinuity Example
x1 = np.linspace(-2,0.93,400)
x2 = np.linspace(1.07,3,400)
y1 = x1**2
y2 = x2**2

plt.plot(x1,y1,color='r')
plt.plot(x2,y2,color='r')
plt.scatter(1,1,s=70,facecolor='none',edgecolor='k')
plt.show()
```

4: Removeable Discontinuities: this is like a hole discontinuity, except that the point that would normally be gone in a hole discontinuity is somewhere else such that the limit and functional value at the pointâ€™s x-coordinate are not the same.

```python
# Removeable Discontinuity Example
x1 = np.linspace(-2,0.93,400)
x2 = np.linspace(1.07,3,400)
y1 = x1**2
y2 = x2**2

plt.plot(x1,y1,color='r')
plt.plot(x2,y2,color='r')
plt.scatter(1,1,s=70,facecolor='none',edgecolor='k')
plt.scatter(1,6,s=70,facecolor='k',edgecolor='k')
plt.show()
```

## Section 3: Left/Right Continuity

Left continuity is achieved when the functional value at $x=a$ is the same as the limit as $x$ approaches $a$ from the left side.

Right continuity is achieved when the functinoal value at $x=a$ is the same as the limit as $x$ approaches $a$ from the right side.

The example used for the jump discontinuity illustrates right continuity, and the first example used for the infinite discontinuity illustrates left discontinuity.

## Section 4: Intervals of Continuity

Intervals of continuity are parts of the domain in which the function is continuous. Discontinuities in the domain are said to not be part of the intervals of continuity.

Let's look at the intervals of continuity of $f(x) = \frac{x^2 + 4}{x(x^2 - 9)}$. Discontinuity occurs when the denominator is equal to 0, so one could say that the function is continuous so long as the denominator is not 0. 

To find the discontinuous parts of the domain, we check to see when the denominator equals 0, that is, $x(x^2-9) = 0$. This occurs when $x = 0, -3, 3$.

We see that the intervals of continuity of $f(x)$ are $(-\infty, -3), (-3, 0), (0,3),$ and $(3,\infty)$. Below is a graph to better illustrate the discontinuities.



Now let's look at the intervals of continuity of $f(x) = \frac{x-4}{x^2 - 16}$. The denominator can be rewritten as $x^2-16 = (x-4)(x+4)$, so we see that $f(x) = \frac{x-4}{(x-4)(x+4)}$. Our first order of business would be to find the discontinuities, which are at $x=4, -4$. We can cancel out the term $(x-4)$ from the top and bottom to get that $f(x) = \frac{1}{x+4}$, but keep in mind that there is still a discontinuity at $x=4$. 

Even though the graph looks continuous around $x=4$, there is still a discontinuity at $x=4$. This is an example of a hole discontinuity.

Thus, the intervals of continuity are $(-\infty,-4), (-4,4)$, and $(4,\infty)$

```python
# Graph
x1 = np.linspace(2,3.94,400)
x2 = np.linspace(4.06,6,400)
y1 = 1/(x1+4)
y2 = 1/(x2+4)

plt.plot(x1,y1,color='r')
plt.plot(x2,y2,color='r')
plt.scatter(4,0.125,s=70,facecolor='none',edgecolor='k')
plt.show()
```

Let's do one more example involving a piecewise function. Take, for example,

$$f(x) =  \begin{cases} 
      \frac{2}{x} & x< 2 \\
      x^2 &  x \geq 2
   \end{cases}
$$

We can see that $f(x)$ is not continuous since the limits are different on both sides: the limit on the left is 1, while the limit on the right is 4. We can also see that $f(x)$ does not exist at $x=0$ (I'm sure you know why). However, $f(x)$ is right continuous since the function is defined at $x=2$ along the function on the right ($x^2$). Since the only discontinuous portions of the graph are at $x=0$, $x=2$, we can conclude that the interval of continuity is $x \in \mathbb{R}, x \neq 0,2$ (meaning that $f(x)$ is continuous when $x$ is any real number excluding 0 and 2).

```python
# Graph (also, the function is right continuous)
x1 = np.linspace(-2,2,400)
x2 = np.linspace(2,6,400)
y1 = 2/x1
y2 = x2**2

plt.plot(x1,y1,color='r')
plt.plot(x2,y2,color='r')
plt.ylim(-7,30)
plt.scatter(2,1,s=70,facecolor='none',edgecolor='k')
plt.scatter(2,4,s=70,facecolor='k',edgecolor='k')
plt.show()
```
