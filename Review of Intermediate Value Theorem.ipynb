```python
import numpy as np
import matplotlib.pyplot as plt
```

## Section 1: What is IVT?

The Intermediate Value Theorem (IVT) can be used to determine if a function $f(x)$ with a continuous interval $[a, b]$ has a point c in the interval $[a, b]$ such that $f (c)$ is between $f(a)$ and $f(b)$, in which $f(a) \neq f(b)$. This is useful for finding roots as well. Your best bet for finding the interval in question is to guess and check; select some random x-values and plug them in to your function.

Below are some examples illustrating IVT.

```python
# Example 1: Given f(x) = 3^x - 4, find an interval that contains the point c where f(c) = 10 using IVT.
t = np.linspace(0,4,500)
y = 3**t - 4
y1 = 10 + 0*t
plt.plot(t,y)
plt.plot(t,y1)
plt.vlines(2,-3,77)
plt.vlines(3,-3,77)
plt.scatter(2,5,s=70,facecolor='k',edgecolor='k')
plt.scatter(3,23,s=70,facecolor='k',edgecolor='k')
plt.show()
```

In the first example, we are trying to find the existence of a point $c$ such that $f(c)=10$ given $f(x)=3^x-4$. The graph illustrates that the point $c$ is between $x=2$ and $x=3$, as illustrated by the horizontal orange line intersecting with the blue curve. To prove this, we will find $f(2)$ and $f(3)$. We see that $f(2) = 5$, and $f(3)=23$. 

Since we know that $5<10$, $23>10$, $f(2) = 5$, and $f(3)=23$, we can conclude that there exists a point $c$ between $x=2$ and $x=3$ such that $f(c)=10$. This is because $f(2)<f(c) = 10<f(3)$.

```python
# Example 2: Given f(x) = x^4 - x^2 - 4x - 7, find an interval that contains the point c where f(c) = -9 using IVT.
t = np.linspace(-1,2,500)
y = t**4 - t**2 - 4*t - 7
y1 = -9+0*t
plt.plot(t,y)
plt.plot(t,y1)
plt.vlines(0,-11,-3)
plt.vlines(1,-11,-3)
plt.scatter(0,-7,s=70,facecolor='k',edgecolor='k')
plt.scatter(1,-11,s=70,facecolor='k',edgecolor='k')
plt.show()
```

In the second example, we can see that the point $c$ is between $x=0$ and $x=1$. To prove this, we will find $f(0)$ and $f(1)$. We see that $f(0) = -7$ and $f(1) = -11$. 

Since we know that $-7>-9$, $-11<-9$, $f(0) = -7$, and $f(1) = -11$, we can conclude that there exists a point $c$ between $x=0$ and $x=1$ such that $f(c) = -9$. This is because $f(0)>f(c)=-9>f(1)$.

## Section 2: Another Application of IVT

IVT can be applied to the more specific example of finding x-intercepts. The theory behind this is that if $f(a)>0$ and $f(b)<0$, then there exists a point $c$ in $[a,b]$ such that $f(c)=0$. It can also be applied when $f(a)<0$ and $f(b)>0$. When you have a continuous interval that goes from negative to positive, you will be absolutely sure to find 0 in between the two intervals.

```python
# Example: Given f(x) = 5^x + 7x, find an interval that contains the solution of f(x) using IVT.
t = np.linspace(-2,1,300)
y = 5**t + 7*t 
y1 = 0 + 0*t
plt.plot(t,y)
plt.plot(t,y1)
plt.vlines(-1,-15,12)
plt.vlines(0,-15,12)
plt.scatter(-1,-6.8,s=70,facecolor='k',edgecolor='k')
plt.scatter(0,1,s=70,facecolor='k',edgecolor='k')
plt.show()
```

In this example, we can see that the solution $c$ is between $x=-1$ and $x=0$. To prove this, we will find $f(-1)$ and $f(0)$. We can see that $f(-1) = -6.8$ and $f(0) = 1$.

Since we know that $0$ is between $-6.8$ and $1$, and we know that $f(-1) = -6.8$ and $f(0) = 1$, we can conclude that there exists a solution $c$ in between $[-1,0]$ such that $f(c)=0$.

In the case of applying IVT to find zeroes, the only thing that matters is that sign change between two different x-values, so an interval of $[-2,1]$ or $[-10,5]$ would also suffice since there will be a sign change there.

## Section 3: Limitations of IVT

Let's take the function $f(x) = (x-4)(x+3)(x-1)(x+2)$ and try to see if there is an x-intercept in the interval $[-5,5]$. Of course, there are four solutions, but assuming that we do not know where said solutions are, we will apply IVT. We can see that $f(-5) = 324$ and $f(5) = 224$. It is clear that zero is not between 224 and 324, so we can not conclude that there is a solution to $f(x)$ in the interval $[-5,5]$ with IVT.

One must be careful when picking the domain for IVT. This is why it is important to guess and check for multiple values of $x$.

```python
# Example 1: Given f(x) = (x-4)(x+3)(x-1)(x+2), find a solution to f(x) using IVT in the interval [-5,5].
t = np.linspace(-5,5,500)
t1 = np.linspace(-6,6,400)
y = (t-4)*(t+3)*(t-1)*(t+2)
y1 = 0 + 0*t
b1 = 175+0*t1
b2 = 350+0*t1
plt.plot(t,y)
plt.plot(t,y1)
plt.plot(t1,b1,color='r')
plt.plot(t1,b2,color='r')
plt.vlines(-6,175,350,color='r')
plt.vlines(6,175,350,color='r')
plt.vlines(-5,-61,324)
plt.vlines(5,-61,324)
plt.scatter(-5,324,s=70,facecolor='k',edgecolor='k')
plt.scatter(5,224,s=70,facecolor='k',edgecolor='k')
plt.show()
```

In the red box containing $f(-5)$ and $f(5)$, we can see that 0 is not between those values. The solutions to $f(x)$ are outside of those values, implying that there does not exist a solution to $f(x)$ in the interval $[-5,5]$ when using IVT.

Note that the red box is added for clarity.

------------

Let's try one more example to drive it home. Let's try to see if there is a solution to $f(x) = 5^x - 6x^2 - x^4$ in the interval $[-1,3]$ using IVT.

We can see that $f(-1) = -6.8$ and $f(3)=-10$. We know that 0 is not in between -10 and -6.8, so we do not know if there is a solution to $f(x) = 5^x - 6x^2 - x^4$ in the interval $[-1,3]$, even if the graph below clearly shows that there are two.

In the red box containing $f(-1)$ and $f(3)$, we can see that 0 is not in between the two values. The solutions to $f(x)$ are outside of those values, implying that there does not exist a solution to $f(x)$ in the interval $[-5,5]$ when using IVT.

```python
# Example 2: Given f(x) = 5^x - 6x^2 - x^4, find a solution to f(x) using IVT in the interval [-1,3].
t = np.linspace(-1,3,500)
t1 = np.linspace(-1.5,3.5,400)
y = 5**t - 6*t**2 - t**4
y1 = 0 + 0*t
b1 = -13+0*t1
b2 = -4+0*t1
plt.plot(t,y)
plt.plot(t,y1)
plt.plot(t1,b1,color='r')
plt.plot(t1,b2,color='r')
plt.vlines(-1.5,-13,-4,color='r')
plt.vlines(3.5,-13,-4,color='r')
plt.vlines(-1,-20,2)
plt.vlines(3,-20,2)
plt.scatter(-1,-6.8,s=70,facecolor='k',edgecolor='k')
plt.scatter(3,-10,s=70,facecolor='k',edgecolor='k')
plt.show()
```

## Section 4: IVT with Two Functions

I am adding this section because I had a lot of students last year that had a lot of difficulty with a question of this calibre on the first midterm.

Let's say we want to prove that there is an intersection of $f(x) = 3^x$ and $g(x) = 0.5(4^x)$, that is, we want to prove that there is a solution such that $f(x)=g(x)$. We can rewrite this as $3^x = 0.5(4^x)$, as well as $3^x - 0.5(4^x) = 0$. We can apply IVT here to look for a solution; what we are looking for is a sign change when testing different x-values.

Let's try using $x=0$, $x=1$, $x=2$, and $x=3$. 

We see that when $x=0$, $3^0 - 0.5(4^0) = 0.5$

When $x=1$, $3^1 - 0.5(4^1) = 1$

When $x=2$, $3^2 - 0.5(4^2) = 1$

When $x=3$, $3^3 - 0.5(4^3) = -5$

The sign change occurs when going from $x=2$ to $x=3$. We can thus conclude that there does exist a point such that $f(x) = 3^x$ and $g(x) = 0.5(4^x)$ intersect, or $f(x)-g(x)=0$.

```python
# A plot of f(x) - g(x), or 3^x - 0.5(4^x)
t = np.linspace(0,3,500)
y = 3**t-0.5*(4**t)
y1 = 0+0*t
plt.plot(t,y)
plt.plot(t,y1)
plt.scatter(0,0.5,s=70,facecolor='k',edgecolor='k')
plt.scatter(1,1,s=70,facecolor='k',edgecolor='k')
plt.scatter(2,1,s=70,facecolor='k',edgecolor='k')
plt.scatter(3,-5,s=70,facecolor='k',edgecolor='k')
plt.show()
```

```python
# A plot of f(x) and g(x)
t = np.linspace(0,3,500)
y1 = 3**t
y2 = 0.5*(4**t)
plt.plot(t,y1,label='f(x)')
plt.plot(t,y2,label='g(x)')
plt.legend()
plt.show()
```

```python
# A closer look of f(x) and g(x)
t = np.linspace(2,3,500)
y1 = 3**t
y2 = 0.5*(4**t)
plt.plot(t,y1,label='f(x)')
plt.plot(t,y2,label='g(x)')
plt.legend()
plt.show()
```
