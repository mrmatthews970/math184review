## Section 1: Please Don't Tell Me There is Trigonometry Involved...

A secant line measures the average slope between two points $a$ and $b$ on a continuous interval of a function $f(x)$. The slope of a secant line is,

$$m_{sec} = \frac{f(b)-f(a)}{b-a}$$

The closer the points $a$ and $b$ are, the closer we get to the actual slope. Let's look at the secant lines of $f(x) = x^2 + 4x$. Since the minimum/vertex of $f(x)$ is at $x=-2$, we can say that the slope $m$ at $x=-2$ is 0.

Let's make estimations starting at $a=-2$, adjusting $b$ so that it gets closer to $a$. We will start with $b=-1$.

We can see that $f(-2) = -4$, and $f(-1) = -3$, so $m_{sec} = \frac{f(-1)-f(-2)}{-1-(-2)} = \frac{-4-(-3)}{1} = \frac{1}{1} = 1$.

As you can tell, this estimation is, er, um, trash. Let's try an estimation with $b=-1.9$. We can see that $f(-1.9) = -3.99$, so $m_{sec} = \frac{f(-1.9) - f(-2)}{-1.9-(-2)} = \frac{-3.99-(-4)}{0.1} = \frac{0.01}{0.1} = 0.1$.

Ok, looking better. Let's try the estimation with $b=-1.99$. We can see that $f(-1.99)=-3.9999$, so $m_{sec} = \frac{f(-1.99)-f(-2)}{-1.99-(-2)} = \frac{-3.9999-(-4)}{0.01} = \frac{0.0001}{0.01} = 0.01$.

Now we're getting there! I will put some graphs below to illustrate each estimation. The black dots are the points, the blue curve is $f(x)$, and the red line is the secant line.

```python
# b=-1
x = np.linspace(-2.2,-0.9,400)
y = x**2 + 4*x
z = x-2

plt.plot(x,y,color='b')
plt.plot(x,z,color='r')
plt.scatter(-2,-4,s=70,facecolor='k',edgecolor='k')
plt.scatter(-1,-3,s=70,facecolor='k',edgecolor='k')
plt.show()
```

```python
# b=-1.9
x = np.linspace(-2.2,-0.9,400)
y = x**2 + 4*x
z = 0.1*x-3.8

plt.plot(x,y,color='b')
plt.plot(x,z,color='r')
plt.scatter(-2,-4,s=70,facecolor='k',edgecolor='k')
plt.scatter(-1.9,-3.99,s=70,facecolor='k',edgecolor='k')
plt.show()
```

```python
# b=-1.99
x = np.linspace(-2.2,-0.9,400)
y = x**2 + 4*x
z = 0.01*x-3.98

plt.plot(x,y,color='b')
plt.plot(x,z,color='r')
plt.scatter(-2,-4,s=70,facecolor='k',edgecolor='k')
plt.scatter(-1.99,-3.9999,s=70,facecolor='k',edgecolor='k')
plt.show()
```

```python
# Closer look of b=-1.99
x = np.linspace(-2.05,-1.95,400)
y = x**2 + 4*x
z = 0.01*x-3.98

plt.plot(x,y,color='b')
plt.plot(x,z,color='r')
plt.scatter(-2,-4,s=70,facecolor='k',edgecolor='k')
plt.scatter(-1.99,-3.9999,s=70,facecolor='k',edgecolor='k')
plt.show()
```

```python
# The actual "secant" line
x = np.linspace(-3,0,400)
y = x**2 + 4*x
z = 0*x-4

plt.plot(x,y,color='b')
plt.plot(x,z,color='r')
plt.scatter(-2,-4,s=70,facecolor='k',edgecolor='k')
plt.show()
```

## Section 2: Application of Secant Lines

One application of secant lines is for physics. You can use secant lines to calculate average velocities or speeds given a certain displacement or distance over a certain amount of time (wait, when was this a physics class!?)

Let's say that a person runs around a 400m track (which amounts to 4km) in 20 minutes (one-third of an hour). If we want to find out the average SPEED of the person in kilometres per hour, we can take the distance (4km) and divide it by the time (13h) to get 12km/h. 

However, the average VELOCITY is different. Since the person runs around the track 10 times, they just end up starting and ending on the same spot, which amounts to a 0km displacement. Therefore, the average velocity is 0km/h. Talk about running in circles...
