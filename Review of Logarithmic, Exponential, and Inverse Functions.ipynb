## Section 1: Exponential Functions

Exponential functions follow the form $y = a(b^{cx-d}) + g$. 

$a$ is the vertical expansion factor, where $a>1$ means that the function is stretched vertically, $0<a<1$ means that the function is compressed vertically, and a negative $a$ means that the function is flipped over the x-axis.

$c$ is the horizontal expansion factor, where $b>1$ means that the function is compressed vertically, $0<b<1$ means that the function is stretched vertically, and a negative $b$ means that the function is flipped over the y-axis.

$d$ is the horizontal shift factor, where a positive $d$ means that the function is moved to the right, and a negative $d$ means that the function is shifted to the left.

$g$ is the vertical shift factor, where a positive $g$ means that the function is moved upwards, and a negative $g$ means that the function is shifted downwards.

$b$ is the exponential base. In this class, $b$ will always be greater than 0.

When $b>1$, the graph gets larger as $x$ increases, and when $0<b<1$, the graph gets smaller as $x$ increases.

The domain for an exponential function is $x \in \mathbb{R}$ (all of $x$), and the range varies depending on the sign of $a$. If $a>0$, then the range is $y>g$, and if $a<0$, then the range is $y<g$.

#### Here are some rules about exponents to jog your memory

- $a^{m+n} = a^m \cdot a^n$
- $(ab)^n = a^n \cdot b^n$
- $\sqrt[n]a = a^\frac{1}{n}$
- $\frac{1}{a^n} = a^{-n}$
- $(a^n)^m = a^{mn}$

Below are some graphs of exponential functions. Note that the axes are not labelled (way to go Matthew -_-. Minus ten points for home team).

```python
# Exponential Functions
x = np.linspace(-1,1,500)
y = 3**x
z = (2/5)**x

plt.plot(x,y, label='y = 3^x')
plt.plot(x,z, label='y=(2/5)^x')
plt.legend()
plt.show()
```

## Section 2: Logarithmic Functions

Logarithmic functions follow the form $y = a\log_b(cx-d) + g$. They are kind of like exponentials, but flipped. You can think of exponents and logarithms as polar opposites.

Note that logarithms can only be computed if $cx - d >0$. Thus, it follows that logarithmic functions are defined when $cx - d>0$. The range for a logarithmic function is $y \in \mathbb{R}$.

The common logarithm (base 10) is given as $\log x$, and the natural logarithm (base $e$, which is Euler's number) is $\ln x$.

#### Here are some rules about logarithms to jog your memory

- $\log_a (bx) = \log_a(x) + \log_a(b)$
- $\log_a (\frac{x}{b}) = \log_a(x) - \log_a(b)$
- $\log_a(x^n) = n\log_a(x)$
- $\log_a(x) = \frac{\log x}{\log a} = \frac{\ln x}{\ln a}$

#### While we're at it, here are some no-nos about logarithms to jog your memory and possibly traumatize you >:)

- $\log_a(x+b) \neq \log_a(x) + \log_a(b)$
- $\log_a(x-b) \neq \log_a(x) - \log_a(b)$

There is no way to simplify a logarithm function in the form $\log_a(x+b)$. Just leave it at that point.

Below are some graphs of logarithmic functions. Feel free to chastise me for the lack of axes... again

```python
# Logarithmic Functions

x = np.linspace(0.00001,5.01,500)
y = np.log2(x)
z = np.log10(x)
a = np.log(x)
plt.plot(x,z, label='y=log(x), common logarithm')
plt.plot(x,a, label='y=ln(x), natural logarithm')
plt.plot(x,-y, label='y=log_(1/2)(x)')
plt.ylim(-2.5,2.5)
plt.legend()
plt.show()
```

## Section 3: Inverse Functions

Say we have a function $f(x)$. The inverse of said function ($f^{-1}(x)$) is like $f(x)$, but flipped around the line $y=x$. Basically, all the $x$ values become the $y$ values, and vice versa. Inverse functions only exist if $f(x)$ is one-to-one, that is, for every value $y$ of the range, there is ONLY ONE value $x$ in the domain such that $y = f(x)$ (ooh, look at me being all technical). 

A way to test for "one-to-one-ness" is to do the horizontal line test: if you can get more than one x-value when you lay the ruler horizontally, then your function is not one-to-one. You need to limit the domain such that one-to-one-ness is satisfied.

Here are some functions that are one-to-one, there will be graphs to show this quality of "one-to-one-ness".

- $y = x^3 + 4$
- $y = \frac{1}{4x - 5}$
- $y = -e^x$
- $y = 10\sqrt x$

Here are some functions that are not one-to-one. Graphs will be below.

- $y = \cos(4x)$
- $y = x^4$
- $y = x^3 - 6x^2$

```python
# One-to-One Functions

x = np.linspace(-2,2,400)
x1 = np.linspace(0,4,400)
x2 = np.linspace(1,1.5,400)

y = x**3 + 4
z = 1/(4*x2-5)
a = -np.exp(x)
b = 10*np.sqrt(x1)
plt.plot(x,y, label='y=x^3 + 4')
plt.legend()
plt.show()
```

```python
plt.plot(x2,z, label='y=1/(4x-5)')
plt.legend()
plt.ylim(-20,20)
plt.show()

# Note that the vertical line at x=1.25 is the discontinuity. 
```

```python
plt.plot(x,a,label='y=-e^x')
plt.legend()
plt.show()
```

```python
plt.plot(x1,b,label='y=10sqrt(x)')
plt.legend()
plt.show()
```

```python
# Not One-to-One Functions
x = np.linspace(-3,3,400)
y = np.cos(4*x)
z = x**4
a = x**3 - 6*x**2
plt.plot(x,y, label='y=cos(4x)')
plt.legend()
plt.show()
```

```python
plt.plot(x,z, label='y=x^4')
plt.legend()
plt.show()
```

```python
plt.plot(x,a, label='y=x^3-6x^2')
plt.legend()
plt.show()
```

Let's try solving for an inverse of a function. Let's take $f(x) = 2(x+1)^5 - 1$. Note that $y$ and $f(x)$ are interchangeable.

Our first step is to replace $x$ with $y$ and vice versa. In this case, we get $x = 2(y+1)^5 - 1$. We now try and isolate for $y$ as a function of $x$, in which we call this new $y$ $f^{-1}(x)$. I will show the steps and annotate along the way.

- $f(x) = 2(x+1)^5 - 1$
- $y = 2(x+1)^5 - 1$ (replaced $f(x)$ with $y$)
- $x = 2(y+1)^5 - 1$ (replaced $x$ with $y$ and vice versa)
- $x+1 = 2(y+1)^5$ (added 1)
- $\frac{x+1}{2} = (y+1)^5$ (divided by 2)
- $\sqrt[5]{\frac{x+1}{2}} = y+1$ (took the fifth root)
- $y = f^{-1}(x) = \sqrt[5]{\frac{x+1}{2}} - 1$ (subtracted 1. Since $y$ is isolated, we can now replace $y$ with $f^{-1}(x)$)

Another one: let's try $f(x) = \frac{3x + 5}{x-4}$. Steps are below.

- $f(x) = \frac{3x + 5}{x-4}$
- $y = \frac{3x + 5}{x-4}$ (replaced $f(x)$ with $y$)
- $x = \frac{3y + 5}{y-4}$ (replaced $x$ with $y$ and vice versa)
- $x(y-4) = 3y + 5$ (multiplied by $(y-4)$)

- $xy - 3y = 4x+5$ (subtracted $3y$)
- $y(x-3) = 4x+5$ (multiplied out $y$ from the term $xy - 3y$)
- $y = f^{-1}(x) = \frac{4x+5}{x-3}$ (divided by $(x-3)$)

Finally, let's find the inverse of a function that is not one-to-one. Let's take $f(x) = \frac{1}{9}(x + 3)^2 - 16$.

This is a parabola. Parabolas are not one-to-one functions since there can be two x-values that share one y-value, so in order to make this function one-to-one, we will need to do a domain restriction. 

There are two parts that we need to worry about: the part left of the vertex, and the part right of the vertex. However, since the parabola is in vertex form ($y = a(x-b)^2 + c$), we can see that the vertex of the parabola is at (-3,-16). The left and right parts of the parabola stem from the vertex. 

We can now divide the parabola into two functions, $f_1(x)$ (the part left of the vertex) and $f_2(x)$ (the part right of the vertex). We can see that,

- $f_1(x) = \frac{1}{9}(x + 3)^2 - 16, x \leq -3$
- $f_2(x) = \frac{1}{9}(x + 3)^2 - 16, x \geq -3$

We will find the inverse of $f_1(x)$. Why that only? It's because I'm lazy. Steps are below.

- $f_1(x) = y = \frac{1}{9}(x + 3)^2 - 16, x \leq -3$
- $x = \frac{1}{9}(y + 3)^2 - 16$. Keep in mind that $y \leq -3$.
- $x+16 = \frac{1}{9}(y + 3)^2$
- $9(x+16) = (y+3)^2$
- $\pm \sqrt{9(x+16)} = y+3$
- $y = \pm \sqrt{9(x+16)} - 3$. At this point, we know that $y \leq -3$, so we will need to take the negative version of the square root.

Thus, $f^{-1}_1(x) = -\sqrt{9(x+16)} - 3$.
