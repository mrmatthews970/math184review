```python
import numpy as np
import matplotlib.pyplot as plt
```

## Section 1: What is Differentiability?

Differentiability of a function $f(x)$ is satisfied when these two conditions are met:

- $\lim_{x\to a}f(x) = f(a)$ (the rule of continuity)
- $\lim_{x\to a}f'(x) = f'(a)$ (the rule of continuity along the derivative)

To paraphrase, the function must be continuous, and so does its derivative.

## Section 2: When is a Function Not Differentiable?

A function is not differentiable when either one of these three things happen:

1. The function is discontinuous somewhere (see the Review of Continuous Functions)

2. There is a sharp bump or cusp in the function

3. The graph looks like it is going completely upwards

Below are some non-differentiable examples for 2 and 3.

```python
# Sharp Bump (look at x=0)
t = np.linspace(-1,1,300)
y = np.abs(t)
plt.plot(t,y)
plt.show()
```

```python
# A More Subtle Sharp Bump (look at x=0)
t = np.linspace(-1,1,300)
y = np.abs(t) + 2*t
plt.plot(t,y)
plt.show()
```

```python
# Cusp (look at x=0)
t = np.linspace(-0.5,2,300)
y = np.sqrt(np.sqrt(np.sqrt(np.abs(t))))
plt.plot(t,y)
plt.show()
```

```python
# Upwards Portion (look at x=0)
t = np.linspace(-1,1,300)
y = np.cbrt(np.cbrt(np.cbrt(t)))**5
plt.plot(t,y)
plt.show()
```
