## Section 1: Background

For business problems, we are trying to find the amount of profit that will be made given our revenue (money made from sales) and cost. Our profit is calculated as $P = R - C$, where $P$ is profit, $R$ is revenue, and $C$ is cost. 

If our profit is positive, then our revenue outweighs our cost, and there is net gain. 

If our profit is negative, then our cost outweighs our revenue, and there is net loss. 

If our profit is exactly zero, then we have a "break-even point", but if the number of products at the "break-even point" is not an integer (that is, not 0, 1, 2, 3, etc.), then the break-even point does not exist.

----
Our revenue can be given as the demand ($q$) multiplied by the price ($p$). Unitwise, demand is in units, and price is in dollars per unit. We can say that $R = pq$.

Our cost usually has a fixed cost, which is constant, and the cost per unit made is multiplied by the demand ($q$).

----
Marginal profit is the derivative of our profit function. If the marginal profit is 0, then we have reached our maximum profit. If the marginal profit is positive, then more profit is being made, that is, profit increases. If the marginal profit is negative, then profit decreases.

Marginal revenue is the derivative of our revenue function, and marginal cost is the derivative of our cost function, as one would assume.

## Section 2: Example 1

Here is one problem that we will run through:

A bakery is selling cakes. When the cakes are priced at \\$15, then 750 units are sold, and for every \\$1 increase, 125 less cakes are sold. However, the cakes come at a cost: the fixed cost for the ingredients is \\$3500, and every cake costs \\$5 to make.

We need to find the slope for the price. We can use our information to see that when the bakery sells cakes at \\$15, then 750 units are sold, and an increase of \\$1 drops the demand by 125, that is, when the cakes are priced at \\$16, then 625 units are sold. Our points to work with are $(15,750)$ and $(16,625)$. We can now find our slope.

$m = \frac{y_2-y_1}{x_2-x_1} = \frac{625-750}{16-15} = -125$

We can work with point-slope form to find our demand function (note, point-slope form is written as $(y-y_1) = m(x-x_1)$). In our demand and unit model, we replace $y$ with $q$, as well as $x$ with $p$. We can use our point $(15,750)$ and our slope $m=-125$ to see that, 

$(q-750)=-125(p-15)$, or $q=-125p + 2625$.

However, for the sake of finding the revenue, we need to convert our demand function into a price function. We get that, $q = -125p+2625$, so we arrange our terms to get that $p = -\frac{q}{125} + 21$. Multiplying this price $p$ by the demand $q$, we see that $R(q) = pq = q(-\frac{q}{125}+21)$.

Our cost is more straightforward: we know that our fixed cost is \\$3500, and the cost of the cakes is \\$5 per unit, so our cost function is $C(q) = 3500 + 5q$.

Thus, our profit function is $P(q) = R(q) - C(q) = (-\frac{q^2}{125} + 21q) - (3500+5q) = -\frac{q^2}{125} + 16q - 3500$.

-----------

If we want to find the break-even points, then we need to find where $P(q)=0$. We can use the quadratic formula for the profit function to find the points. Using it, we get that,

$q = \frac{-16 \pm \sqrt{16^2 - 4(-\frac{1}{125})(-3500)}}{2(-\frac{1}{125}} = \frac{125(-16 \pm \sqrt{256-112})}{-2} = \frac{125(16 \pm 12)}{2} = 1000 \pm 750$

We see that the break-even points are at $q = 250, 1750$. Since those are integers, then the break-even points exist. This means that the revenue equals the cost when $q=250$ and $q=1750$. In terms of price of the products, one would break even when the products cost either \\$19 or \\$7 (hint: use your price function $p = -\frac{q}{125}+21$).

---------

If we want to find the maximum profit, we can do this two ways. First, we can find the x-coordinate of the vertex, that is, $q = \frac{-b}{2a}$. Secondly, we can find the x-intercept of the marginal profit. Using both methods, we see that for maximal profit, $q$ must be 1000, that is, when 1000 cakes are sold, the most profit is made. In terms of price of the products, one would produce the most revenue when the products cost \\$13.

Below is a graph of the profit, cost, and revenue functions. The black function is the revenue, the blue function is the cost, and the green function is the profit. The green line at $y=0$ is used to denote break-even points. The black dots and lines also denote the break-even points.

```python
# Graph of R, C, P
t = np.linspace(0,2500,10000)
f1 = -t**2/125 + 21*t
f2 = 3500 + 5*t
f3 = f1-f2
z = 0*t
plt.plot(t,f1,label = 'R(q)', color='k')
plt.plot(t,f2,label = 'C(q)', color='b')
plt.plot(t,f3,label = 'P(q) and 0', color='g')
plt.plot(t,z, color='g')
plt.scatter(250,4750,s=70,facecolor='k',edgecolor='k')
plt.scatter(1750,12250,s=70,facecolor='k',edgecolor='k')
plt.vlines(250,-15000,15000)
plt.vlines(1750,-15000,15000)
plt.legend()
plt.show()
```

## Section 3: Example 2

Here is another problem:

A music shop is selling a new model of guitars. They estimate that if the guitars are priced at \\$1500, then 50 of them would sell per week, and if the guitars are priced at \\$1350, then 52 of them would sell per week. However, buying the guitars from the manufacturer comes with a fixed cost of \\$7500, as well as a shipping cost of \\$300 for each guitar.

First, we need to find out the revenue function. We see that we have two points to work with: $(1500,50)$ and $(1350,52)$. We see that the slope is $m = \frac{52 - 50}{1350 - 1500} = -\frac{1}{75}$. Using the point $(1500,50)$ and slope-point form, we can see that $q-50 = -\frac{1}{75}(p-1500)$, or $q = -\frac{1}{75}p + 70$. 

Rearranging to isolate for $p$, we see that $p = -75q + 5250$. Given that $R = pq$, we can see that $R(q) = -75q^2 + 5250q$, thus giving us our revenue function.

Our cost function $C(q) = 300q+7500$. We can now find our profit function, which is $P(q)=-75q^2+4950q-7500$.

-----
To find the break-even points, we use the quadratic formula. We can see that the break-even points are at $q=1.552, 64.448$. Since these are not integers, we can say that there are no break-even points.

---
To find the maximum profit, we will use the marginal profit function. We see that $M(q)=-150q+4950$. The maximum profit occurs when $M(q)=0$, which is achieved when $q=33$. To paraphrase, the most profit can be made when 33 guitars are sold. In terms of price of products, one would expect to produce the most revenue when the products cost \\$2775.

Below is the graph of the profit, cost, and revenue functions. The colour scheme is the same as before.

```python
# Graph of R, C, P
t = np.linspace(0,75,1000)
f1 = -75*t**2 + 5250*t
f2 = 7500 + 300*t
f3 = f1-f2
z = 0*t
plt.plot(t,f1,label = 'R(q)', color='k')
plt.plot(t,f2,label = 'C(q)', color='b')
plt.plot(t,f3,label = 'P(q) and 0', color='g')
plt.plot(t,z,color='g')
plt.legend()
plt.show()
```

## Section 4: Price Changes: Adapted from Example 2

What I would like to now demonstrate is the behaviour of the profit when costs change. Let's say that we price the guitars at \\$2250. What we want to know is how the profit changes if the price increases a little bit. 

First, what we need to find is the number of units that would sell if a guitar were priced at \\$2250. We use $q = -\frac{1}{75}p + 70$ and plug in $p=2250$. We get that $q=40$. We see that $R(40) = 70500$. 

If we increase the price to, let's say \\$2325, we see that $q=39$, and we get that $R(39) = 71475$. We see that if we slightly increased the price of the guitar from \\$2250, then more profit would be made.

--------
Now, let's try pricing the guitars at \\$750 and seeing what happens when we decrease the price. We use $q=-\frac{1}{75}p + 70$ and plug in $p=750$. We get that $q=60$, and $R(60)=19500$. 

If we decrease the price to \\$675, we see that $q=61$, and we get that $R(61) = 15375$. We see that if we slightly decreased the price of the guitar from \\$750, then less profit would be made.

--------
Finally, let's try pricing the guitars at \\$3750 and deciding if it would be more economically sound to increase or decrease the price of the guitars, that is, we need to decide if either increasing or decreasing the price will lead to more profits. When $p=3750$, $q=20$. We get that $R(20)=61500$. 

First, we shall increase the price to $p=3825$. We get that $q=19$ and $R(19)=59475$. We see that the profit decreases, so increasing the price is not an economically sound option. 

Next, we shall decrease the price to $p=3675$. We get that $q=21$, and $R(21)=63375$. We see that the profit increases, so decreasing the price is the economically sound option. Thus, decreasing the price of the guitars from \\$3750 would produce more profit.
