As you may remember, the definition of a derivative is this:

$$ f'(x) = \lim_{h \to 0}\frac{f(x+h)-f(x)}{h}$$

This can also be rewritten as:

$$ f'(x) = \lim_{h \to 0}\frac{f(x+h)-f(x)}{(x+h)-x}$$

At $x=a$, we see that:

$$ f'(a) = \lim_{h \to 0}\frac{f(a+h)-f(a)}{(a+h)-a}$$

We can also use the secant line definition, where $a$ gets extremely close to $b$. We can see that:

$$ f'(a) = \lim_{b \to a}\frac{f(b)-f(a)}{b-a}$$ or, $$f'(b)=\lim_{a \to b}\frac{f(b)-f(a)}{b-a}$$

These may seem pretty obvious, but you will need to keep these ideas in mind for the questions we are about to cover.

## Question 1

We want to compute the limit $\lim_{h \to 0}\frac{(2+h)^6 - 16(2+h)^2}{h}$, which represents $f'(a)$ at some point $x=a$. 

The first step is to look at the definition of the derivative at a specific point, namely $f'(a)=\lim_{h \to 0}\frac{f(a+h)-f(a)}{h}$. 

We can see that as $h$ approaches 0, then $(2+h)^6 - 16(2+h)^2$ approaches 0, so we get that $f(2)=0$. We can also see that we let $a=2$, so we can see that $f(2+h)=(2+h)^6 - 16(2+h)^2$. Replacing $2+h$ with $x$, we can see that $f(x)=x^6-16x^2$.

The next step is to calculate $f'(a)$ at $a=2$. We know $f(x)$, so all we need to do is derive $f(x)$ and plug in $x=2$ (note that we said that $x=2+h$ and $a=2$. When we let $h$ approach 0, we see that $x$ approaches 2, which just so happens to be $a$). We see $f'(x)=6x^5-32x$, and $f'(a)=f'(2)=128$.

## Question 2

We want to compute the limit $\lim_{x \to -1}\frac{2^{(x+1)} - 1}{x+1}$ this time.

Once again, we take a look at the definition of the derivative, which is normally $f'(a)=\lim_{h \to 0}\frac{f(a+h)-f(a)}{h}$. We can see that as $x$ approaches -1, $2^{(x+1)}$ approaches $2^0$ or 1, thus $2^{(x+1)}-1$ approaches 0, so we get that $f(-1)=0$. Remembering that we let $x=-1$, we get that $f(x)=2^{(x+1)}$.

Now, we can compute $f'(x)$ at $x=-1$. We see that $f'(x)=2^{(x+1)}\ln(2)$, so $f'(-1)=2^0\ln(2)=\ln(2)$.

## Question 4

This last one will be extra painful...

We want to compute the limit $\lim_{x \to 2}\frac{x^2-4}{f(x)-f(2)}$ given that $f'(2)=4$. Using the close secant line definition ($ f'(a) = \lim_{b \to a}\frac{f(b)-f(a)}{b-a}$), we can see that $f'(2)=4=\lim_{x\to 2}\frac{f(x)-f(2)}{x-2}$. We can rewrite $\lim_{x \to 2}\frac{x^2-4}{f(x)-f(2)}$ as $\lim_{x \to 2}\frac{(x-2)(x+2)}{f(x)-f(2)} = \lim_{x \to 2}(x+2)\frac{(x-2)}{f(x)-f(2)}$. 

Since $4=\lim_{x\to 2}\frac{f(x)-f(2)}{x-2}$, we get that $\lim_{x \to 2}\frac{x-2}{f(x)-f(2)}=\frac{1}{4}$. Applying this, we get that $\lim_{x \to 2}(x+2)\frac{(x-2)}{f(x)-f(2)}=\lim_{x \to 2}(x+2)(\frac{1}{4}) = 4(\frac{1}{4})=1$.

## Question 5

We now want to compute the limit $\lim_{x \to -5} \frac{(x^2-25)(x^2+8x+15)}{x(f(-5)-f(x))^2}$ given that $f'(-5)=6$. Using the close secant line definition ($f'(b)=\lim_{a \to b}\frac{f(b)-f(a)}{b-a}$), we can see that $f'(-5) = 6 = \lim_{x \to -5}\frac{f(-5)-f(x)}{-5-x} = -\lim_{x \to -5}\frac{f(-5)-f(x)}{5+x}$. To make it simpler, we will say that $-6=\lim_{x \to -5}\frac{f(-5)-f(x)}{5+x}$.

We can rewrite $\lim_{x \to -5} \frac{(x^2-25)(x^2+8x+15)}{x(f(-5)-f(x))^2}$ as $\lim_{x \to -5} \frac{(x+5)(x-5)(x+5)(x+3)}{x(f(-5)-f(x))^2}$, or $\lim_{x \to -5} \frac{(x+5)^2(x-5)(x+3)}{x(f(-5)-f(x))^2}$. 

We know that $-6=\lim_{x \to -5}\frac{f(-5)-f(x)}{5+x}$, so we get that $\lim_{x \to -5}\frac{5+x}{f(-5)-f(x)} = -\frac{1}{6}$. 

Rewriting, we get that $\lim_{x \to -5} \frac{(x+5)^2(x-5)(x+3)}{x(f(-5)-f(x))^2} = \lim_{x \to -5}\frac{(x-5)(x+3)}{x}(\frac{5+x}{f(-5)-f(x)})^2 = \lim_{x \to -5}\frac{(x-5)(x+3)}{x}(-\frac{1}{6})^2 = (\frac{-80}{-5})(\frac{1}{36}) = \frac{4}{9}$.
